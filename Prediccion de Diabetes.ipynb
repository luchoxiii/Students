## Descripción del Dataset

Este dataset contiene información sobre la salud de personas, con el objetivo de predecir la presencia de diabetes. Las columnas del dataset representan diferentes características relacionadas con la salud de los individuos:

- **Pregnancies**: Número de veces que la persona ha estado embarazada.
- **Glucose**: Nivel de glucosa en sangre de la persona.
- **BloodPressure**: Presión arterial (en mmHg).
- **SkinThickness**: Grosor de la piel en milímetros.
- **Insulin**: Nivel de insulina en sangre.
- **BMI**: Índice de masa corporal de la persona.
- **DiabetesPedigreeFunction**: Función que representa la historia familiar de diabetes.
- **Age**: Edad de la persona en años.
- **Outcome**: Variable objetivo que indica si la persona tiene diabetes (1) o no (0).

El dataset se utiliza para predecir si una persona tiene diabetes en función de las características mencionadas.

### Estadarizacion & Desarrollo del modelo

```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = df['Outcome']
```

```python
print(X)
print(Y)
```

##### Test & Train

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

#### Making a Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('La persona no tiene diabetes')
else:
  print('La persona tiene diabetes')
```
