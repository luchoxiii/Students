### Descripción

Este es un dataset del sonar de un submarino, donde las variables numéricas representan lo que detecta el radar, y el último valor indica la posibilidad de detectar una piedra o mina.

### Comienzo de analisis & modelo

```python
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
